```python
from glob import glob

def dir_to_df(path, **kwars):
    """Modified version of Joshuas read function. Takes directory wih cvs's and concatenates them."""
    path = Path(path)
    all_files = sorted(path.glob('./*.csv'))
    return pd.concat(pd.read_csv(filename, **kwargs) for filename in all_files)
```
