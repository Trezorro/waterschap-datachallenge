```python
df = pd.read_excel(r'./data/sewer_model/20180717_dump riodat rioleringsdeelgebieden_matched_to_rainfall_locations.xlsx')
df = df.drop(df.index[0:8])
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df.reset_index(drop= True)
df1 = df['RWZI']=="R.W.Z.I. Den Bosch"
df2 = df[df1]
df2 = df2.drop(columns=['Eigendom', 'Afv cap',  'Begin', 'ONP', 'Type riol st', 'Afv wijze'])
DenBoschPompen = df2
```

```python
DenBoschPompen['Afvoer naar'] = DenBoschPompen['Afvoer naar'].fillna(value= 0)
DenBoschPompen = DenBoschPompen[DenBoschPompen['Naam / lokatie'] != 'Geb. 31 Treurenburg']
DenBoschPompen = DenBoschPompen[DenBoschPompen['Naam / lokatie'] != 'Geb. 16 Paleiskwartier']
Eindpompen = DenBoschPompen[DenBoschPompen['Afvoer naar']==0]
DenBoschPompen[DenBoschPompen['Naam / lokatie'] != 'Geb. 31 Treurenburg']
```

```python
Bokhoven = ['HER-BOK-KER']
Haarsteeg = ['HEU-HAA-KER']
Helftheuvelweg = ['HER-HER-KRU']
Maaspoort = ['HER-HER-MAA']
OudenEngelseweg = ['HER-HER-CNR']
Rompert = ['HER-HER-NOO']

pompen = [Bokhoven, Haarsteeg, Helftheuvelweg, Maaspoort, OudenEngelseweg, Rompert]
df = DenBoschPompen.set_index('Code')
#df.loc['HER-HER-GEE']
```

```python
pompdict = defaultdict(list)
for index, row in df.iterrows():
    
    trace = [row.name]
    node = df.loc[row.name]
    while node['Afvoer naar'] !=0:
        try:
            node = df.loc[node['Afvoer naar']]
        except:
            print("Not found:", node['Afvoer naar'])
            trace.append(None)
            break
        trace.append(node.name)
    pompdict[trace[-1]] = list(set(pompdict[trace[-1]] + trace))
        
```
